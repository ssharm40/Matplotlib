**Legend()**-        For giving the label name

```python
# If you want to give lable name then you must call the legend method
```

```python
from matplotlib import pyplot as plt

x1=[1,2,3]
y1=[2,5,7]

x2=[0.5,2,4]
y2=[1,5,8]


plt.title("Two Data in a Single Frame")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.plot(x1,y1,label='First')
plt.plot(x2,y2,label='Second')
plt.legend()
plt.show()
```

```python
#If you want to increse the width of line---------- linewidth


x1=[1,2,3]
y1=[2,5,7]

x2=[0.5,2,4]
y2=[1,5,8]


plt.title("Two Data in a Single Frame")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.plot(x1,y1,label='First', linewidth=4)
plt.plot(x2,y2,label='Second', linewidth=3)
plt.legend()
plt.show()
```

**Scatter Plot**

```python
#If you want to print in scatter format---------means dotted format
#In place of plot use Scatter


x1=[1,2,3]
y1=[2,5,7]

x2=[0.5,2,4]
y2=[1,5,8]


plt.title("Two Data in a Single Frame")
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.scatter(x1,y1,label='First', linewidth=4)
plt.scatter(x2,y2,label='Second', linewidth=3)
plt.legend()
plt.show()
```
