```python
# Matplotlib - is a visualization tool.
# for Graphs
# Line graph, bar,stacked, pie graph and many more

%matplotlib inline
 
from matplotlib import pyplot as plt
import numpy as np
```

```python
# For x,y simple graph


x=[2005,2006,2007,2008,2009,2010,2011,2012]
y=[20,35,48,40,49,58,69,32]

plt.plot(x,y)
plt.show()
```

```python
# Name to Graph
# xlabel- give name to x-label
# ylabel- give name to y-label


x=[2005,2006,2007,2008,2009,2010,2011,2012]
y=[20,35,48,40,49,58,69,32]

plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Year Wise Production')

plt.plot(x,y)
```

```python
#Grid function
# Backside me grid krne k lie

x=[2005,2006,2007,2008,2009,2010,2011,2012]
y=[20,35,48,40,49,58,69,32]

plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Year Wise Production')

plt.grid(True)

plt.plot(x,y)
```

```python
# To increse the width of line
#linewidth = number
#To change color -   'r'


plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Year Wise Production')
plt.grid(True)

plt.plot(x,y,'r', linewidth=6)
plt.show()
```

```python
# Now two manipulation in a singe graph
#legend
# To recognize the line of graph, we use legend. If you want to use Label then you must need to give a label in plot section



x=np.arange(2006,2021,2)
y1=[20,35,48,40,49,58,69,32]
y2=[25,35,52,69,50,48,84,32]

plt.xlabel('Year')
plt.ylabel('Production')

plt.grid(True)

plt.plot(x,y1,'r',label='Delhi',linewidth=5)
plt.plot(x,y2,'b',label='Mumbai',linewidth=5)

plt.legend()

plt.show()
```

# Pie Graph

```python
from matplotlib import pyplot as plt
import numpy as np
```

We want to create a graph for what we do in our 24 hours

sleep-7 hrs    eating-2hrs    walking-2hrs       working-13hrs

```python
#labels=== Name assign krne k lie
#startangle= 90 degree se start
# We take the color of graph as a list in cols

hours=[7,2,2,13]
activities=['sleeping','eating','walking','working']

cols=['c','m','r','b']

#Simple Graph
#plt.pie(hours)

#Now Taking the name of activities, colour
#plt.pie(hours,colors=cols,labels=activities)


#Now taking the startangle
plt.pie(hours,colors=cols,labels=activities,startangle=90)



plt.title('Daily Routine')
plt.show()
```

```python
#Shadow dene k lie -  shadow=True
#Explode - kisi part to special show krne k lie
#autopct-  percetage show krne k lie
#savefing-- To save file

```

```python
#For Shadow

days=[7,2,2,13]
activities=['sleeping','eating','walking','working']
cols=['c','m','r','b']

plt.pie(days, colors=cols, labels=activities, startangle=90, shadow=True)

plt.title('Daily Routine')
plt.show()
```

```python
#Explode - kisi part to special show krne k lie


plt.pie(days, colors=cols, labels=activities, startangle=90, shadow=True, explode=[0,0,0,0.08])
plt.title('Daily Routine')

plt.show()
```

```python
#autopct-  percetage show krne k lie

plt.pie(days, colors=cols, labels=activities, startangle=90, shadow=True, explode=[0,0,0,0.08], autopct='%1.1f')
plt.title('Daily Routine')

plt.legend()
plt.show()
```

```python
#savefing-- To save file

plt.pie(days, colors=cols, labels=activities, startangle=90, shadow=True, explode=[0,0,0,0.08], autopct='%1.1f')
plt.title('Daily Routine')

plt.savefig('Pie1.jpg')
plt.legend()
plt.show()
```

```python
#For Covid Data

pr=[29,11,18,11,21,12]
label=['Loss of taste & smell','Age','Fever','Tiredness','Breathing Problem','Headache']
cols=['b','m','r','b','y','g']
plt.pie(pr, colors=cols, labels=label, startangle=90, shadow=True, explode=[0.049,0,0,0,0,0], autopct='%1.2f')
plt.legend()
plt.subplots_adjust(0.3,.1,1.5,2)

plt.show()
```

```python
# For Adjusting the size of graph

plt.pie(days, colors=cols, labels=activities, startangle=90, shadow=True, explode=[0,0,0,0.08], autopct='%1.1f')
plt.title('Daily Routine')

plt.subplots_adjust(0.3,.1,1.5,2)
plt.show()
```

```python
# To make a black circle in  between- we can also say this a DOUNAT graph
#GCF-- Graphic Configuration
#GCA-- Graphic Artist

days=[7,2,2,13]
activities=['sleeping','eating','walking','working']
cols=['c','m','r','b']

plt.pie(days,colors=cols,labels=activities,startangle=90,shadow=True,explode=[.05,0,0,0],autopct='%1.1f%%')

mcircle=plt.Circle((0,0),0.7,color='white')
p=plt.gcf()
p.gca().add_artist(mcircle)
plt.legend()
plt.title('Daily Routine')
```

# Bar Graph

```python
#xticks use for label ko name dene k lie

x=['Nucelar','Hydro','Gas','Oil','Coal','Biofuel']
enrgy=[5,6,15,22,24,8]
pos=len(x)
pos=np.arange(0,pos,1)

plt.bar(pos,enrgy,color='blue')

plt.xlabel='Energy Source'
plt.ylabel='Energy Output'
plt.title('Energy Output from Different sources')

plt.xticks(pos,x)
plt.show()
```

```python
#Try another one
#Alpha--------For manager the color level


lang=['Python','Java','C','JavaScript','C#']
pos=np.arange(len(lang))
performance=[10,8,3,9,5]
plt.bar(pos,performance,color='green',align='center',alpha=.8)
plt.xticks(pos,lang)
plt.title('Programming Language Performance')
plt.show()
```

```python
#barh------For horizontal Graph---only barh is needed


lang=['Python','Java','C','JavaScript','C#']
pos=np.arange(len(lang))
performance=[10,8,3,9,5]
plt.barh(pos,performance,color='green',align='center',alpha=.8)
plt.xticks(pos,lang)
plt.title('Programming Language Performance')
plt.show()
```

```python
# For Subplot


group=4
priyansh=(90,55,40,65)
tanmay=(85,62,54,20)
fig,ax=plt.subplots()
idx=np.arange(group)
bar_width=.35
r1=plt.bar(idx,priyansh,bar_width,alpha=.8,color='blue',label='Priyansh')
r2=plt.bar(idx+bar_width,tanmay,bar_width,alpha=.8,color='green',label='Tanmay')
plt.xlabel('Person')
plt.ylabel('Score')
plt.title('Score of person')
plt.xticks(idx+bar_width,('A','B','C','D'))
plt.legend()
plt.show()
```
